### Train Operational Readouts

Cuando hablamos de un 'readout' nos referimos a una medición o registro de datos que se obtiene de un sensor, instrumento o sistema durante un experimento, prueba o proceso. En este caso específico, los 'operational readouts' se refieren a las mediciones o registros que se obtienen durante la operación de un sistema o equipo, como puede ser una máquina, un vehículo o cualquier otro tipo de dispositivo que esté en funcionamiento. Estos readouts pueden incluir una variedad de parámetros, como temperatura, presión, velocidad, voltaje, corriente, entre otros, dependiendo del tipo de sistema que se esté monitoreando. Estos datos son cruciales para evaluar el rendimiento del sistema, detectar posibles fallas o anomalías, y optimizar su funcionamiento.

Algo que podemos hacer con esta tabla es verificar si hay valores nulos o faltantes en alguna de las columnas. Esto es importante porque los valores nulos pueden afectar el análisis y los modelos predictivos que se construyan posteriormente. Podemos usar el siguiente código para verificar la presencia de valores nulos:

```python
# Valores nulos totales
# esto representa del total, sabiendo que el total de datos son las filas multiplicadas por las columnas

total_filas = train_operational_readouts.shape[0] * train_operational_readouts.shape[1]
total_nulos = train_operational_readouts.isnull().sum().sum()
porcentaje_nulos = (total_nulos / total_filas) * 100
print(f'Total de valores nulos: {total_nulos}')
print(f'Porcentaje de valores nulos: {porcentaje_nulos:.2f}%')
```

```python
#valores nulos
train_operational_readouts.isnull().sum()
```

Podemos darnos cuenta que la tabla contiene muchos valores nulos pero comparados con el total de datos no representan un porcentaje alto. Ni siquiera el 1% del total de datos son nulos.

Ahora vamos a verificar que en la columna 'vehicle_id' al sacar los valores únicos, estos sean la misma cantidad que el de las otras tablas. 

```python
# valores unicos de vehicle_id
train_operational_readouts['vehicle_id'].nunique()
```

#### imputar valores nulos

```python
# para hacerlo vamos a hacerlo con una imputacion interpolacion temporal, ya que los datos son series de tiempo y la interpolacion puede ser una buena tecnica para llenar los valores nulos
train_operational_readouts = (
    train_operational_readouts
    .sort_values(by=['vehicle_id', 'time_step'])
    .groupby('vehicle_id')
    .apply(lambda group: group.interpolate(method='linear'))
    .reset_index(drop=True)
)
```

```python
# volvemos a verificar los valores nulos
train_operational_readouts.isnull().sum()
```

Todavia quedan unos cuantos valores, los cuales probablemente correspondan a los inicios o finales de las series de tiempo de cada vehiculo, por lo que no se pueden interpolar. Para estos casos, podemos optar por rellenar los valores nulos restantes con el método 'bfill' (backward fill) o 'ffill' (forward fill), dependiendo de la naturaleza de los datos y del contexto del análisis. Estos métodos rellenan los valores nulos utilizando el valor más cercano disponible en la serie temporal, ya sea hacia adelante o hacia atrás.

```python
# hagamos el metodo bfill para los valores nulos restantes
train_operational_readouts = train_operational_readouts.bfill()
train_operational_readouts = train_operational_readouts.ffill()
```

```python
# ahora si verificamos los valores nulos, ya no deberia haber ninguno
train_operational_readouts.isnull().sum().sum()
```

```python
#hacemos de nuevo el analisis estadistico descriptivo para ver si cambio algo
train_operational_readouts.describe()
```

Nos damos cuenta que despues de hacer la imputacion de valores nulos, ya no quedan valores nulos en la tabla. Y que cuando hacemos el analisis estadistico descriptivo, los valores no cambian mucho, lo que indica que la imputacion no afecto significativamente la distribucion de los datos. Pues los valores nulos eran pocos en comparacion con el total de datos.

#### Análisis y visualización de datos

Bien ahora usando esta base limpia podemos empezar nuestro analisis y visualzacion de datos. Podemos darnos cuenta que cada columna corresponde a un sensor diferente, y que cada fila corresponde a una medicion en un tiempo determinado para un vehiculo especifico. Los sensores estan agrupados en diferentes categorias, por ejemplo, el sensor 167, esta dividido en 167_0, 167_1, 167_2, etc. Esto indica que el sensor 167 tiene varias mediciones o sub-sensores asociados a el. A estas subdivisiones las llamaremos 'bins'.

```python
# vamos a intentar ver cuantos sensores hay y cuantos bins tiene cada sensor
# vamos a descartar obvio la columna de vehicle_id y time_step
sensor_columns = train_operational_readouts.columns.difference(['vehicle_id', 'time_step'])
sensors = {}
for col in sensor_columns:
    sensor_id = col.split('_')[0]
    if sensor_id not in sensors:
        sensors[sensor_id] = 0
    sensors[sensor_id] += 1
# imprimimos el resultado
for sensor_id, bin_count in sensors.items():
    print(f'Sensor {sensor_id} tiene {bin_count} bins')
    
```

Para que una variable sea considerada 'histogrameable', debe tener mas de un bin en este caso. Por tanto solo podemos considerar histogrameables:

```python
# vamos a ver aquellas variables con mas de 1 bin
histogrameable_sensors = {sensor_id: bin_count for sensor_id, bin_count in sensors.items() if bin_count > 1}
# imprimimos el resultado
for sensor_id, bin_count in histogrameable_sensors.items():
    print(f'Sensor {sensor_id} tiene {bin_count} bins')
```

Entonces por ejemplo si usamos como ejemplo el vehiculo #2, podemos sacar los histogramas de las variables histogrameables para ese vehiculo en particular.

```python
# histograma de la varible 167 para el vehiculo 2
# sin saber cuales son todos los bins de la variable 167, vamos a sacarlos dinamicamente

import matplotlib.pyplot as plt

vehicle_id = 2
bins_167 = [col for col in sensor_columns if col.startswith('167_')]

# Filtramos datos del vehículo
df_vehicle = train_operational_readouts[train_operational_readouts['vehicle_id'] == vehicle_id]

# Calculamos la suma de cada bin
freqs = df_vehicle[bins_167].sum()

# Colores distintos para cada barra
colors = plt.cm.tab20(np.arange(len(bins_167)))

# Graficamos
plt.figure(figsize=(10, 6))
plt.bar(freqs.index, freqs.values, color=colors, edgecolor='black')

plt.title(f'Suma de valores de los bins del sensor 167 para el vehículo {vehicle_id}')
plt.xlabel('Bins del sensor 167')
plt.ylabel('Suma de valores')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# vamos a hacer lo mismo pero para la variable 459 que tiene 20 bins
import numpy as np
vehicle_id = 2
bins_459 = [col for col in sensor_columns if col.startswith('459_')]
# Filtramos datos del vehículo
df_vehicle = train_operational_readouts[train_operational_readouts['vehicle_id'] ==
    vehicle_id]
# Calculamos la suma de cada bin
freqs = df_vehicle[bins_459].sum()
# Colores distintos para cada barra
colors = plt.cm.tab20(np.arange(len(bins_459)))
# Graficamos
plt.figure(figsize=(10, 6))
plt.bar(freqs.index, freqs.values, color=colors, edgecolor='black')
plt.title(f'Suma de valores de los bins del sensor 459 para el vehículo {vehicle_id}')
plt.xlabel('Bins del sensor 459')
plt.ylabel('Suma de valores')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

Bien ahora con las otras 8 variables numericas podemos hacer un analisis similar, estas son las variables que no tienen sub-sensores o bins, las cuales serian:

```python
# variables sin bins
# vamos a ver cuales son las variables que no tienen bins, es decir, que solo tienen una columna
non_histogrameable_sensors = [sensor_id for sensor_id, bin_count in sensors.items() if bin_count == 1]
print('Variables sin bins:', non_histogrameable_sensors)
```

```python
import matplotlib.pyplot as plt
import numpy as np

vehicle_id = 2

# Filtrar datos del vehículo y ordenar por time_step
df_vehicle = train_operational_readouts[train_operational_readouts['vehicle_id'] == vehicle_id].copy()
df_vehicle = df_vehicle.sort_values('time_step')

# Detectar variables sin bins desde el diccionario sensors
non_histogrameable_sensors = [sensor_id for sensor_id, bin_count in sensors.items() if bin_count == 1]

# Ajustar nombres al formato real del DataFrame (añadir "_0")
non_bin_vars = [f"{sensor}_0" for sensor in non_histogrameable_sensors if f"{sensor}_0" in df_vehicle.columns]

# Crear paleta de colores
colors = plt.cm.get_cmap('tab10', len(non_bin_vars))

plt.figure(figsize=(12, 7))
for i, var in enumerate(non_bin_vars):
    plt.plot(df_vehicle['time_step'], df_vehicle[var] / 1e8,
             label=var, color=colors(i), marker='o', linestyle='-')

plt.title(f'Variables sin bins (detectadas automáticamente) para el vehículo {vehicle_id}')
plt.xlabel('Time Step')
plt.ylabel('Valores (x1e8)')
plt.grid(True)
plt.legend(loc='upper left', bbox_to_anchor=(1, 1))
plt.tight_layout()
plt.show()
```

Adicionalmente podemos realizar tambien un grafico de correlacion entre estas variables numericas sin bins. 

```python
import matplotlib.pyplot as plt
import seaborn as sns

vehicle_id = 2

df_vehicle = train_operational_readouts[train_operational_readouts['vehicle_id'] == vehicle_id].copy()
df_vehicle = df_vehicle.sort_values('time_step')

# Detectar variables sin bins desde el diccionario sensors
non_histogrameable_sensors = [sensor_id for sensor_id, bin_count in sensors.items() if bin_count == 1]

# Ajustar nombres al formato real del DataFrame (añadir "_0")
non_bin_vars = [f"{sensor}_0" for sensor in non_histogrameable_sensors if f"{sensor}_0" in df_vehicle.columns]

# Subset del DataFrame con solo esas variables
df_subset = df_vehicle[non_bin_vars]

# Filtrar columnas constantes (std = 0)
df_subset = df_subset.loc[:, df_subset.std() != 0]

# Calcular matriz de correlación
corr_matrix = df_subset.corr()

# Graficar heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5,
            vmax=1.0, vmin=-1.0)

plt.title(f'Correlación dinámica entre variables sin bins para el vehículo {vehicle_id}')
plt.tight_layout()
plt.show()
```

```python
# Si no quiero filtrar por un vehiculo en particular, puedo hacer la correlacion para todos los vehiculos juntos

import matplotlib.pyplot as plt
import seaborn as sns

# Usar todo el DataFrame (sin filtrar por vehicle_id)
df_all = train_operational_readouts.copy()
df_all = df_all.sort_values('time_step')

# Detectar variables sin bins desde el diccionario sensors
non_histogrameable_sensors = [sensor_id for sensor_id, bin_count in sensors.items() if bin_count == 1]

# Ajustar nombres al formato real del DataFrame (añadir "_0")
non_bin_vars = [f"{sensor}_0" for sensor in non_histogrameable_sensors if f"{sensor}_0" in df_all.columns]

# Subset del DataFrame con solo esas variables
df_subset = df_all[non_bin_vars]

# Filtrar columnas constantes (std = 0)
df_subset = df_subset.loc[:, df_subset.std() != 0]

# Calcular matriz de correlación (ahora sobre todos los vehículos)
corr_matrix = df_subset.corr()

# Graficar heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5,
            vmin=-1, vmax=1)

plt.title('Correlación dinámica entre variables sin bins (todos los vehículos)')
plt.tight_layout()
plt.show()
```

Es que decir esta columna in_study_repair no es como tal un indicador numerico si no booleano en donde 1 indica que hubo una reparacion y 0 indica que no hubo reparacion. Por lo que no tiene sentido hacer un analisis estadistico descriptivo de esta columna. Podemos hacer una distribucion de pastel para ver la proporción de reparaciones y no reparaciones en el conjunto de datos.

```python
# grafica pastel de la columna in_study_repair
import matplotlib.pyplot as plt

# Etiquetas y valores
labels = ['No Repair', 'Repair']
sizes = train_tte['in_study_repair'].value_counts().values
colors = ['navy', 'darkred']

# Función para mostrar porcentaje y cantidad
def autopct_format(pct, allvals):
    absolute = int(round(pct/100.*sum(allvals)))
    return f"{pct:.1f}%\n({absolute})"

plt.figure(figsize=(6, 6))
plt.pie(
    sizes,
    labels=labels,
    colors=colors,
    autopct=lambda pct: autopct_format(pct, sizes),
    startangle=140,
    textprops={'color':"white", 'weight':'bold'}
)
plt.title('Proporción de Reparaciones en el Conjunto de Datos')
plt.show()
```

Ademas podemos tener una distribucion de tiempo de observacion en el 'x' seria el tiempo de observacion y en el 'y' la cantidad de vehiculos que estuvieron en ese tiempo de observacion.

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))

# Histograma separado por clase (hue)
sns.histplot(
    data=train_tte,
    x='length_of_study_time_step',
    hue='in_study_repair',       # separa 0 y 1
    bins=50,
    palette={0: 'navy', 1: 'darkred'},  # azul para 0, rojo para 1
    alpha=0.6,
    edgecolor='white',
    multiple='stack'             # apila las barras
)

plt.title('Distribución del Tiempo de Observación por Estado de Reparación')
plt.xlabel('Tiempo de Observación (length_of_study_time_step)')
plt.ylabel('Número de Vehículos')
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
