### Funcion para cargar los datos desde un archivo CSV

```python
import os

def load_data(file_path):
    if not os.path.exists(file_path):
        raise FileNotFoundError(f"Archivo no encontrado: {file_path}")
    return pd.read_csv(file_path)
```
